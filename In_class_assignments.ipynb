```python
class Person:
  def __init__(self, first_name, last_name):
    self.first = first_name
    self.last = last_name

  def display(self, formal=False):
    if formal:
      return f"{self.last}, {self.first}"
    else:
      return f"{self.first}"

  def __str__(self):
    return f"Person: {self.first} {self.last}"
```

```python
# Create an instance of the Person class
my_person = Person("kique", "loser")

# Display informally
print(f"Informal display: {my_person.display()}")

# Display formally
print(f"Formal display: {my_person.display(formal=True)}")

# Using the __str__ method (implicit call with print)
print(my_person)
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of class Greeting"

name = Greeting("Kique", "chud")
print(name)
name.display()
name.display(formal=True)
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    if newWage <= 0:
      print("Invalid wage")
    else:
      self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def display_info(self):
    print(f"Employee: {self.name}")
    print(f"Hourly wages: ${self.__hourly_wage}")
    print(f"Hours worked: {self.hours}")
    print(f"Total Pay: ${self.calculate_pay()}")

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}"

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours_worked, bonus):
    super().__init__(name, hourly_wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
      base_pay = super().calculate_pay()
      return base_pay + self.bonus

  def display_info(self):
    super().display_info()
    print(f"Bonus: ${self.bonus}")

print("Employee Information:")
employee = Employee("Kique ", 20, 40)
employee.display_info()

print("Manager Information:")
manager = Manager("Zack", 30, 40, 500)
manager.display_info()
```
