```python
sales = float(input("enter sales value ($):"))
Tax_rate= float(input("enter tax rate (%):"))
cogs = 0.40 * sales
gross_profit = sales - cogs
tax = gross_profit * (Tax_rate/100)
net_income = gross_profit - tax

print(f"Sales: {sales}$")
print(f"Cost Of Goods Sold: {cogs}$")
print(f"gross profit : {gross_profit}$")
print(f"tax: {tax}$")
print(f"net income: {net_income}$")
```

```python
x = int(input("enter number here: "))
y = int(input("enter number here: "))
z = (input("enter operator: "))

if z =="+":
  answer = x + y
elif z == "-":
  answer = x - y
elif z == "*":
  answer = x * y
elif z == "/":
  answer = x / y
else:
  print("invalid operator")
print(answer)
```
