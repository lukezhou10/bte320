```python
sales = float(input("enter sales value ($):"))
Tax_rate= float(input("enter tax rate (%):"))
cogs = 0.40 * sales
gross_profit = sales - cogs
tax = gross_profit * (Tax_rate/100)
net_income = gross_profit - tax

print(f"Sales: {sales}$")
print(f"Cost Of Goods Sold: {cogs}$")
print(f"gross profit : {gross_profit}$")
print(f"tax: {tax}$")
print(f"net income: {net_income}$")
```

```python
import random

while True:
  player = input("Enter rock, paper, or scissors (or 'quit' to stop): ").lower()
  if player == 'quit':
    print("Thanks for playing!")
    break

  if player not in ['rock', 'paper', 'scissors']:
    print("Invalid input. Please enter rock, paper, or scissors.")
    continue

  computer_move = random.choice(['rock','paper','scissors'])
  print(f"Computer chose: {computer_move}")

  if player == computer_move:
    print("It's a tie!")
  elif (
      (player == 'rock' and computer_move == 'scissors') or
      (player == 'paper' and computer_move == 'rock') or
      (player == 'scissors' and computer_move == 'paper')
  ):
    print("You win!")
  else:
    print("Computer wins!")
```
