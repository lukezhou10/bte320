```python
sales = float(input("enter sales value ($):"))
Tax_rate= float(input("enter tax rate (%):"))
cogs = 0.40 * sales
gross_profit = sales - cogs
tax = gross_profit * (Tax_rate/100)
net_income = gross_profit - tax

print(f"Sales: {sales}$")
print(f"Cost Of Goods Sold: {cogs}$")
print(f"gross profit : {gross_profit}$")
print(f"tax: {tax}$")
print(f"net income: {net_income}$")
```

```python
x = int(input("enter number here: "))
y = int(input("enter number here: "))
z = (input("enter operator: "))

if z =="+":
  answer = x + y
elif z == "-":
  answer = x - y
elif z == "*":
  answer = x * y
elif z == "/":
  answer = x / y
else:
  print("invalid operator")
print(answer)
```

```python
ticket_price = 10
defaultaudience = 20
print("Ad$\tProfit")
for admoney in range(0,201,25):
  revenue = (defaultaudience + round(2 * admoney ** 0.5)) * ticket_price
  costs = 200 + admoney
  profit = revenue - costs
  print(f"{admoney} \t {profit}")
```

```python
print("Hours Fee")
Hours = 1
while Hours < 9:
  Fee = 5 + (2.5 * Hours)
  if Fee < 10:
    Fee = 10
  elif Fee > 20:
    Fee = 20
  Hours += 1
  print(f"{Hours}     ${Fee}")
```
