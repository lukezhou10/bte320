```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Numerical Operator: ")

if op == "+":
  res = num1 + num2
elif op =="-":
  res = num1 - num2
elif op =="*":
  res = num1 * num2
elif op =="/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "%":
  res = num1 % num2
elif op == "**":
  res = num1 ** num2
else:
  print("Invalid operator")

print(f"result: {res}")
```
