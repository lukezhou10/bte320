```python
email = "!! Click on the following:  to ..."
is_urgent = "URGENT"
has_www= "www."

if (is_urgent in email) and (has_www in email):
  print("phishing!")
if (is_urgent in email) and not (has_www in email):
  print("spam!")
if not (is_urgent in email) and (has_www not in email):
  print("legit!")
```
